# Maximum Margin Hyperplane

## Margin of a hyperplane to a set of points

$$
\Large \text{Given points $\{x_1, ..., x_{N}\}$ and a plane $H(w, w_0)$, then the margin is given by:}
$$

$$
\Large \text{$M(w, w_0) = min_{x \in \{x_1, ..., x_N\}} d(x, w, w_0) = min_{x \in \{x_1, ..., x_N\}} \frac{|w^{T}x + w_0|}{||w||} $}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
# Parameters
N = 5

data = np.random.randn(N, 2) + [1, 1]
X = data[0:,0:1]
y = data[0:,1:2]

X_continues = np.linspace(0, 3, 1000)
```

```python
# y = -x - 1
hyperplane = -X_continues + 1
w = np.array([1, 1])
w0 = -1
```

```python
plt.scatter(X, y, s=20, color='blue', label='Data')
plt.plot(X_continues, hyperplane, color='green', label='Hyperplane')
plt.legend()
plt.show()
```

```python
# Calculate L2 norm and distance to hyperplane
norm = lambda v: np.sqrt(sum(x**2 for x in v))
dist = lambda x, w, w0: (np.dot(w, x) + w0) / norm(w)
```

```python
def M(data, w, w0) -> float:
    distances = []
    for i, x in enumerate(data):
        distance = dist(x, w, w0)
        distances.append((distance, i))
    return min(distances)
```

```python
_, data_i = M(data, w, w0)
print(f"Margin = {data[data_i]}")
```
